```python
# grid is printed as bins along with bin/column number for ease of visualization.
def print_grid(grid):
    for i in range(ROWS):
        print('| ' + ' | '.join(grid[i]) + ' |  ')

    # Print column numbers below the grid
    print('  ' + '   '.join(str(i) for i in range(COLUMNS)))
```

```python
# function parameters: game grid, disc name: Human / Computer, column / bin number where the disc has to be dropped
# checks if the selected bin can accomodate a disc. Search begins from bottom-up. 
# In case there is the bin is already full, same status is reported for calling function
def drop_disc(grid, disc, column):
    # looping in reverse to place the disc in the last empty row of the given column
    disc_dropped=False
    for i in range(ROWS-1,-1,-1):
        if grid[i][column]==' ':
            grid[i][column]=disc
            disc_dropped=True
            print(f"disc placed in {i}th row  of {column} column")
            break;
    print_grid(grid)
    return disc_dropped
```

```python
def is_grid_column_full(grid, column):
    if grid[0][column] == ' ':
        return False
    else:
        return True
```

```python
# target is to check if one column has an empty cell, then play continues
def is_grid_full(grid):
    grid_has_space = False
    for col in range(COLUMNS):
        if is_grid_column_full(grid, col)== False:
            grid_has_space = True
            break
    return not grid_has_space           
        
```
